## Crear un DataFrame: `.DataFrame`

```python
import pandas as pd
df = pd.DataFrame({"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]})
print('\n df:')
print(df)
```

## Concatenar las filas de 2 DataFrame con `.concat`

```python
import pandas as pd

df1 = pd.DataFrame({"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]})

print('\n df1:')
print(df1)

df2 = pd.DataFrame({"a" : [7, 8, 9],
"b" : [10, 11, 12],
"c" : [13, 14, 15]})

print('\n df2:')
print(df2)

df_concat = pd.concat([df1,df2])

print('\n df_concat:')
print(df_concat)
```

## Ordenar registros con `.sort_values`

```python
df_users = pd.DataFrame({
    "nombre" : ['Ana', 'José', 'Mariela'],
    "edad" : [75, 31, 22]
    })

## De menor a mayor:
print('\n De menor a mayor, utilizando la columna "edad" ')
print(df_users.sort_values('edad'))

## De mayor a menor:
print('\n De mayor a menor, utilizando la columna "edad" ')
print(df_users.sort_values('edad', ascending=False))


## Orden alfabético de la a -> z
print('\n De menor a mayor, utilizando la columna "nombre" ')
print(df_users.sort_values('nombre'))

## Orden alfabético de la a -> z
print('\n De mayor a menor, utilizando la columna "nombre" ')
print(df_users.sort_values('nombre', ascending=False))
```

## Selecionar registros `.head()` y `.tail()`

```python
import pandas as pd

df_users = pd.DataFrame({
    "nombre" : ['Ana', 'José', 'Mariela','Martín','Gabriel','Micaela','Nicolás'],
    "edad" : [75, 31, 22,56,78,98,23]
    })

print("\n Los 5 primeros valores")
print(df_users.head()) # equivale a los 5 primeros valores


print("\n Los 2 primeros valores")
print(df_users.head(2)) 

print("\n Los 3 útlimos valores")
print(df_users.tail(3))
```

## Selecciona y ordena los N registros de mayor a menor dependiendo de la columna que seleciones `df.nlargest(N, 'value’)`

```python
import pandas as pd

df_users = pd.DataFrame({
    "nombre" : ['Ana', 'José', 'Mariela','Martín','Gabriel','Micaela','Nicolás'],
    "edad" : [75, 31, 22,56,78,98,23],
    "pais": ['Argentina','Francia','Croacia','Marruecos','Portugal','Grecia','Colombia' ]
    })

df_users.nlargest(3, 'edad')
```

## Selecciona y ordena los N registros de menor a mayor dependiendo de la columna que seleciones `df.nsmallest(N, 'value’)`

```python
import pandas as pd

df_users = pd.DataFrame({
    "nombre" : ['Ana', 'José', 'Mariela','Martín','Gabriel','Micaela','Nicolás'],
    "edad" : [75, 31, 22,56,78,98,23],
    "pais": ['Argentina','Francia','Croacia','Marruecos','Portugal','Grecia','Colombia' ]
    })

df_users.nsmallest(3, 'edad')
```

### Cuenta la cantidad de registros en el DataFrame: `len(df)` y `df_users.shape`
  



```python
import pandas as pd

df_users = pd.DataFrame({
    "nombre" : ['Ana', 'José', 'Mariela','Martín','Gabriel','Micaela','Nicolás'],
    "edad" : [75, 31, 22,56,78,98,23],
    "pais": ['Argentina','Francia','Croacia','Marruecos','Portugal','Grecia','Colombia' ]
    })

print('\n Cantidad filas en el Dataframe utilizando "len()"')
print(len(df_users))
print('\n Cantidad filas en el Dataframe utilizando ".size[0]"')
print(df_users.shape[0])
print('\n Cantidad columnas en el Dataframe utilizando ".size[1]"')
print(df_users.shape[1])
```

### Cuenta la cantidad de registros con valor único: `df['w'].value_counts()`

```python
import pandas as pd
df_example = pd.DataFrame({
    "letra" : ['O', 'E', 'U', 'E', 'U', 'a', 'O', 'E', 'O', 'U', 'a', 'U', 'a', 'U', 'E'],
    "numero" : [4, 6, 5, 9, 4, 9, 5, 6, 9, 5, 12, 3, 5, 12, 3]
    })

print('df_example: ')
print(df_example)

print('\nNúmero de veces en que cada letra se repite en la columna "letra"')
print(df_example['letra'].value_counts())
```

### Gráfico de dispersión: `df.plot.scatter(x='w',y='h')`
